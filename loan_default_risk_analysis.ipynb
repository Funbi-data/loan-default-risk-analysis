#Define features and Target


```python
x = df.drop('not.fully.paid', axis=1)
y = df['not.fully.paid']
```

X- features(input) Y - target(what we predict)

#Split the Data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

The dataset was split into training and testing sets using a 70-30 ratio. a random state was fixed to ensure reproducibility of results

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

A logistic Regression model wwas trained to predict loan defaul. The model achieved an  accuracy of 84%%, with evaluation metrics such as precision, recall and f1 score used to assess performance

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

The defaulter matrix reaveals, that the model correctly identified 1,609 non defaulters but failed to detect most defaulters, correctly identifying only 5 out of 305. A large number of defaulters 300 were misclassified as safe borrowers, indicating poor performance in detecting high- risk customer.

#Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

Although the model achieved an accuracy of 84.2%, it performed poorly in identifing loan defaulters, with a recall of only 2%. this indicates that the model fails to detect most high-risk customers, making it unstable for real- world financial decision making

#Random Forest Model

o improve the model we imported RandomForestClassifier which is much stronger than Logistic Regression to handle imbalance better and captures complex pattern

```python
y_pre_rf = rf_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy:',accuracy_score(y_test,y_pre_rf))
print(f"Accuracy: {accuracy * 100:.2f}%")
```

The result shows that even after using Random Forest, the model predicts "no default" most of the time. It still fails to detect risky customers, with the recall for defaulters at just 3%. This suggests that the data is imbalanced and needs to be fix.

```python
rf_model = RandomForestClassifier(class_weight ='balanced')
rf_model.fit(x_train,y_train)
```

The dataset was imbalanced , leading to poor detection of defaulters. To adress this, class weighting was applied to improve the model's ability to identify high-risk customers

```python
y_pred_rf=rf_model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred_rf))
```

The dataset was highly imbalanced, leading, leading to poor detection of defaulters. To adress this, SMOTE(SYthetic Minority Oversampling Technique)was applied to balance the classes. This significantly improved the model's ability to identify high-risk customers.

#Handle Class Imbalance with SMOTE

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
x_resampled,y_resampled=smote.fit_resample(x_train,y_train)
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(x_resampled,y_resampled)
```

```python
y_pred_smote = rf_model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred_smote))
```

Initially , the logistc Regression and Random Forest models achieved high acccuracy but failed to identify loan defaulters due to class imbalance. After applying SMOTE, the recall for defaulters improved significantly from 2% to 24%, indicating better detection of high-risk customers. Although accuracy decreased, the model became more prectical for real-world financial risk assessment

# Batch Mode

```python
# Create a batch of applicants (you can expand this)
batch_applicants = pd.DataFrame({
    'credit.policy': [1, 1, 0, 1, 0],
    'int.rate': [0.12, 0.09, 0.15, 0.20, 0.07],
    'installment': [300, 200, 450, 500, 150],
    'log.annual.inc': [10.5, 11.2, 10.0, 9.8, 11.5],
    'dti': [15, 10, 25, 30, 8],
    'fico': [700, 750, 650, 600, 780],
    'days.with.cr.line': [4000, 5000, 3000, 2000, 6000],
    'revol.bal': [10000, 8000, 15000, 20000, 5000],
    'revol.util': [0.45, 0.30, 0.70, 0.85, 0.20],
    'inq.last.6mths': [1, 0, 2, 3, 0],
    'delinq.2yrs': [0, 0, 1, 2, 0],
    'pub.rec': [0, 0, 1, 1, 0]
})
```

# Get Probability Scores

```python
processed_batch_applicants = batch_applicants.copy()
processed_batch_applicants['purpose'] = 'debt_consolidation'
processed_batch_applicants = pd.get_dummies(processed_batch_applicants, columns=['purpose'], drop_first=True)
model_features = x_resampled.columns
missing_cols = set(model_features) - set(processed_batch_applicants.columns)
for c in missing_cols:
    processed_batch_applicants[c] = 0
processed_batch_applicants = processed_batch_applicants[model_features]
probabilities = rf_model.predict_proba(processed_batch_applicants)[:, 1]
batch_applicants['default_probability'] = probabilities
```

# Create Risk Levels

```python
def risk_level(prob):
    if prob < 0.30:
        return 'Low Risk'
    elif prob < 0.60:
        return 'Medium Risk'
    else:
        return 'High Risk'

batch_applicants['risk_level'] = batch_applicants['default_probability'].apply(risk_level)
```

# Add Business Decisions

```python
def decision(risk):
    if risk == 'Low Risk':
        return 'Approve Loan'
    elif risk == 'Medium Risk':
        return 'Manual Review'
    else:
        return 'Reject / High Risk'

batch_applicants['decision'] = batch_applicants['risk_level'].apply(decision)
```

# Final Output

```python
final_output = batch_applicants[
    ['default_probability', 'risk_level', 'decision']
].copy()

final_output['default_probability'] = final_output['default_probability'].round(2)

print(final_output)
```

```python
final_output.sort_values(by='default_probability', ascending=False)
```
